```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../DATA/student_data.csv')
df.head()
```

```python
# Are students with unpaid tuition fees more likely to drop out?
# - Tuition fees up to date: Binary variable indicating whether the tuition fees are up to date (1 for yes, 0 for no).

unpaid_tuition = df[df["Tuition fees up to date"] == 1]
paid_tuition = df[df["Tuition fees up to date"] == 0]


print(len(unpaid_tuition),len(paid_tuition))



df_filtered = df[df['Output'].isin(['Dropout', 'Graduate'])]

plt.figure(figsize=(5, 3))
sns.countplot(data=df_filtered, x='Tuition fees up to date', hue='Output', palette='Set2')

# Labeling
plt.xticks([0, 1], ['Unpaid', 'Paid'])
plt.title('Student Outcome by Tuition Payment Status')
plt.xlabel('Tuition Fees Status')
plt.ylabel('Number of Students')
plt.legend(title='Outcome')
plt.tight_layout()
plt.show()
```

```python
df

# - Debtor: Binary variable indicating whether the individual is a debtor (1 for yes, 0 for no).

debtor = df[df["Debtor"] == 1]
not_debtor = df[df["Debtor"] == 0]


print(len(debtor),len(not_debtor))



df_filtered = df[df['Output'].isin(['Dropout', 'Graduate'])]

plt.figure(figsize=(5, 3))
sns.countplot(data=df_filtered, x='Debtor', hue='Output', palette='Set2')

# Labeling
plt.xticks([0, 1], ['not_debtor', 'debtor'])
plt.title('Student Outcome by debt status')
plt.xlabel('Debt status')
plt.ylabel('Number of Students')
plt.legend(title='Outcome')
plt.tight_layout()
plt.show()
```
